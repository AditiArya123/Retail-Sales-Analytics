# 🛒 Superstore Data Preprocessing using Python (Pandas)

## Objective

This notebook performs data preprocessing and validation on the Superstore dataset before SQL analysis.

The preprocessing workflow includes:

- Loading the raw dataset
- Inspecting the data structure
- Checking for missing values and duplicates
- Standardizing column names
- Validating key business fields
- Generating summary statistics
- Exporting a cleaned dataset for downstream SQL analysis

```python
import pandas as pd
#load the data
df = pd.read_csv("Superstore.csv", encoding="latin1")

df.head()
print(df.shape)
df.info()
```

```python
#checking duplicates rows
duplicate_rows = df.duplicated().sum()
print('Number of duplicate rows:', duplicate_rows)
```

```python
# cheching missing values
missing_values= df.isnull().sum()
print(missing_values)
```

```python
print(df.dtypes)
```

```python
#Convert Dates
df["Order Date"] = pd.to_datetime(df["Order Date"])

df["Ship Date"] = pd.to_datetime(df["Ship Date"])

print(df.dtypes)
```

```python
#Standardize Column Names
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
      .str.replace("-", "_")
)

print(df.columns)
```

```python
#negtaive sales
negative_sales = df[df["sales"] < 0]

print("Negative Sales Records:", len(negative_sales))
```

```python
#invalid discounts
invalid_discount = df[(df["discount"] < 0) | (df["discount"] > 1)]

print("Invalid Discount Records:", len(invalid_discount))
```

```python
#duplicate order_id
# Note:
# Duplicate Order IDs are expected because
# one customer order can contain multiple products.
duplicate_orders = df["order_id"].duplicated().sum()

print("Duplicate Order IDs:", duplicate_orders)
```

```python
#unique categories
print(df["category"].unique())
```

```python
#summary statistics
print(df.describe())
```

```python
# This tells us the dataset size after preprocessing.
print("\nDataset Shape:", df.shape)
```

```python
# Save cleaned dataset
df.to_csv("cleaned_superstore.csv", index=False)

print("Cleaned dataset exported successfully!")
```
